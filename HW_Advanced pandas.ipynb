```python
### Задание 1
df = pd.read_csv('visit_log.csv', sep = ';')
df.head()
def source_type (data):
    if data['traffic_source'] == 'yandex' or data['traffic_source'] == 'google':
        return 'organic'
    elif data['traffic_source'] == 'paid'or data['traffic_source'] == 'email' and data['region'] == 'Russia':
        return 'ad'
    elif data['traffic_source'] == 'paid'or data['traffic_source'] == 'email' and data['region'] != 'Russia':
        return 'other'
    return data['traffic_source']
df['source_type'] = df.apply(source_type, axis = 1)
df.head()
```

```python
### Задание 3
rating  = pd.read_csv('ml-latest-small/ratings.csv')
rating['date'] = pd.to_datetime(rating['timestamp'], unit = 's')
#добавляем столбец, в котором переводим столбец с датой в привычный формат даты
rating_count = rating[['userId','rating']].groupby('userId').count().reset_index()
# создаем датафрейм с количеством оценок по каждому пользователю
date_min = rating[['userId','date']].groupby('userId').min().reset_index()
#создаем датафрейм с минимальной датой выставления оценки
date_max = rating[['userId','date']].groupby('userId').max().reset_index()
#создаем датафрейм с максимальной датой выставления оценки
rating_with_min_date = rating_count[rating_count['rating'] > 100].merge(date_min, how = 'left', on = 'userId')
data = rating_with_min_date.merge(date_max, how = 'left', on = 'userId')
# соединяем 3 датафрейма в один
def life_cycle(df):
    return df['date_y'] - df['date_x']
# пишем функцию вычислябщую жизненый цикл каждого пользователя
data['life_cycle'] = data.apply(life_cycle, axis = 1)
# применяем функцию к нашему датафрейму
data.head()
```
