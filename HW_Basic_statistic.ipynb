```python
#Напишем функцию для разделении температуры/пульса/дыхания на категории: нормальная, повышенная, пониженная

### Тут у меня вопрос. Убила несколько дней в попытке написать универсальную функцию, которую можно было бы применить к нескольким столбцам с доп переменными
### Т.е. def category(age, indicator, нижняя граница для взросылх, верхняя граница для взрослых, нижняя для молодых, верхняя для молодых)
### Что бы в итоге можно было применить одну функцию для создания 3ех новых столбцов с категорией
### где в переменных передавать столбец с возрастом (age), нужный столбец для анализа  - ректальная температура/пульс/частота дыхания (indicator) и показания нормы для взрослых молодых животных
### Как можно это сделать?
def rec_t_category(data):
    if data['Age'] == 1:
        if data['rectal_temperature'] < 37.2:
            return 'пониженная'
        elif data['rectal_temperature'] <= 38.3:
            return 'нормальная'
        elif data['rectal_temperature'] > 38.3:
            return 'повышенная'
        return data['rectal_temperature']
    else:
        if data['rectal_temperature'] < 37.5:
            return 'пониженная'
        elif data['rectal_temperature'] <= 38.9:
            return 'нормальная'
        elif data['rectal_temperature'] > 38.9:
            return 'повышенная'
        return data['rectal_temperature']
def pulse_category(data):
    if data['Age'] == 1:
        if data['pulse'] < 28:
            return 'пониженная'
        elif data['pulse'] <= 44:
            return 'нормальная'
        elif data['pulse'] > 44:
            return 'повышенная'
        return data['pulse']
    else:
        if data['pulse'] < 80:
            return 'пониженная'
        elif data['pulse'] <= 100:
            return 'нормальная'
        elif data['pulse'] > 80:
            return 'повышенная'
        return data['pulse']
def resp_r_category(data):
    if data['Age'] == 1:
        if data['respiratory_rate'] < 10:
            return 'пониженная'
        elif data['respiratory_rate'] <= 24:
            return 'нормальная'
        elif data['respiratory_rate'] > 24:
            return 'повышенная'
        return data['respiratory_rate']
    else:
        if data['respiratory_rate'] < 20:
            return 'пониженная'
        elif data['respiratory_rate'] <= 40:
            return 'нормальная'
        elif data['respiratory_rate'] > 40:
            return 'повышенная'
        return data['respiratory_rate']
stat_horse['rec_t_category'] = stat_horse.apply(rec_t_category, axis = 1)
stat_horse['pulse_category'] = stat_horse.apply(pulse_category, axis = 1)
stat_horse['resp_r_category'] = stat_horse.apply(resp_r_category, axis = 1)
stat_horse.head(10)
```
